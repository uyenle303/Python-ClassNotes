## 3D SURFACE PLOTS

```python
# 3dimensional plots
x = np.linspace(-1, 1, 1000)
y = np.linspace(-1, 1, 1000)

X, Y = np.meshgrid(x,y)
Z = np.exp(-X*X -Y*Y)
fig, ax = plt.subplots(subplot_kw = {'projection': '3d'}, figsize = (8,8))
ax.plot_surface(X,Y,Z, cmap = 'coolwarm')
plt.show()
```

# Exercise: The Mandelbrot set
1. Each complex number $x + iy$ can be represented in the plane as a point of coordinates $(x,y)$.
2. The square of the distance from the center of such point is $x^2 + y^2$  
3. The sequence recursively defined by $x_{n+1} = x_n^2 + c$, given a complex number c, is a sequence of complex numbers which can be plotted in the plane.
4. The sequence either escapes far away from the origin, or stays in a neighborhood of the origin (practically, all the first 50 points of the sequence are less than 2 units away from the origin).
5. The set of c such that the constructed sequence starting from x does not escape from the origin is called the Mandelbrot set. See:  
https://en.wikipedia.org/wiki/Mandelbrot_set
6. Your mission is to use python to draw the Mandelbrot set, or part of it.
7. Look here https://en.wikipedia.org/wiki/Complex_number
   for some info about complex numbers
   

```python
#Question 1 

import numpy as np
import matplotlib.pyplot as plt

# Define a function to check if a point is in the Mandelbrot set
def mandelbrot(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

# Define the parameters for the plot
xmin, xmax = -2, 2
ymin, ymax = -2, 2
width, height = 800, 800
max_iter = 100

# Create a grid of complex numbers
x = np.linspace(xmin, xmax, width)
y = np.linspace(ymin, ymax, height)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Compute the Mandelbrot set
M = np.zeros((height, width))
for i in range(height):
    for j in range(width):
        M[i, j] = mandelbrot(Z[i, j], max_iter)

# Plot the Mandelbrot set
plt.figure(figsize=(10, 10))
plt.imshow(M, extent=(xmin, xmax, ymin, ymax), cmap='hot', interpolation='nearest')
plt.colorbar(label='Iterations')
plt.title('Mandelbrot Set')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.show()

```
