# Exercise:
Given the data:  
group_a = (25, 37, 39, 23, 56)  
group_b = (29, 36, 38, 25, 22)  
Produce the following bar graph:  

  
Check the doc at  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

```python
value = (
    "1",
    "2",
    "3",
    "4",
    "5",
)

groups = {
    "Group A" : np.array([25, 37, 39, 23, 56]),
    "Group B" : np.array([29, 36, 38, 25, 22]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(5)

for boolean, group in groups.items():
    p = ax.bar(value, group, width, label=boolean, bottom=bottom)
    bottom += group

ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.legend(loc="upper left")

plt.axis('off')
plt.show()
```
