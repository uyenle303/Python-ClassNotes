2. Reverse a string. For instance 'abcde' becomes 'edcba'. Try to do this first by writing the string as a list of characters, use the previous exercise to invert this list, and then convert it back to a string. Or try to do this directly with the string.

```python
alpha = ['a', 'b', 'c', 'd', 'e']

def reverse_alpha(char_list):
    char_list.reverse()
    reversed_string = ''.join(char_list)
    return reversed_string

reversed_alpha = reverse_alpha(alpha)
print("Reversed List as String:", reversed_alpha)

```

3. Write a function that multiply two matrices of equal size.

```python
def Mat_mult(m1, m2):
    # Multiplies two matrices of equal size
    n = len(m1)
    result = [[sum(m1[i][k] * m2[k][j] for k in range(n)) for j in range(n)] for i in range(n)]
    return result
```

```python
Mat_mult([[1,2], [3,4]], [[1,0], [0,1]])
```

4. Extact all the integers that are multiple of both 3 and 7 and less or equal to 2000.

```python
ll = [i for i in range(1, 2001)]
for x in ll:
    if x % 3 == 0 and x % 7 == 0:
        print(str(x) + ' is a multiple of both 3 and 7')
    else:
        print(str(x) + ' is not a multiple of both 3 and 7')
```

```python
# Extract integers that are multiples of both 3 and 7 and less than or equal to 2000
multiples_3_and_7 = [num for num in range(1, 2001) if num % 3 == 0 and num % 7 == 0]

# Print the result
print("Integers that are multiples of both 3 and 7 and less than or equal to 2000:")
print(multiples_3_and_7)
```

```python
def guess_number():
    low = 0
    med = 5
    high = 9

    while low <= high:
        guess = (low + med + high) // 3
        response = input(f"Is your number greater than {guess}? (Enter Y for yes, N for no): ").upper()

        if response == 'Y':
            low = guess + 1
        elif response == 'N':
            high = guess - 1
        else:
            print("Please enter Y or N.")

    print(f"Yay! You guessed {low}.")
        
guess_number()
```
