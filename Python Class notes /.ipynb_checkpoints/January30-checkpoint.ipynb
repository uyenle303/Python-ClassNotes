1. A vector of dimension n is a list of n numbers.  
2. A square nxn matrix is a list of n sublists; each sublist is of dimension n.

1. Function that returns the results of a matrix applied to a vector of compatible dimension
2. Function that returns the product of two matrices of same dimension
3. Insert an error message {for instance "Incompatoble dimensions; please try again") if input dimensions are not compatible (one cannot add a vector of length 3 with a vector of length 4, for instance).  

```python
# Function to add two vectors.
def Vect_sum(v1, v2):
    if len(v1) != len(v2):
        raise ValueError("Incompatible dimensions; please try again.")
    # Returns the sum of two vectors of same dimension
    a = [v1[i] + v2[i] for i in range(len(v1))]
    return a
    
```

```python
Vect_sum([1,2,3], [4,5,6]) 
```

```python
def Sum_v(x):
    # Returns sum of elements of a list of numbers
    A = 0
    for y in x:
        A = A + y
    return A   
```

```python
def Dot_prod(v1, v2):
    # Returns dot product of two vectors of same dimension
    n = len(v1)
    a = [v1[i]*v2[i] for i in range(n)]
    return Sum_v(a)
```

```python
def Mat_sum(m1, m2):
    # Returns sum of two matrices of same dimension
    n = len(m1)
    m = [[m1[i][j] + m2[i][j] for j in range(n)] for i in range(n)]
    return m
```

```python
Mat_sum([[1,2], [3,4]], [[1,0], [0,1]])
```

```python
Sum_v([1,2,3])
```
