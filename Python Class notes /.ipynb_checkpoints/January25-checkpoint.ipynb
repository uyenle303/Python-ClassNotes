# Print the first 20 terms of the Fibbonacci sequence
Recall that the Fib. sequence is generated recursively by $x_{n+2} = x_{n+1} + x_n$ starting with $x_0 = 1$ and $x_1 = 1$.

```python
ll = [1,1]
for i in range(18):
    a = ll[-1] + ll[-2]
    ll.append(a)

print(ll)
```

```python
# What if I want only the Fib. numbers less than 1000?
```

```python
limit = 1000
ll = [1,1]
while ll[-1] <= limit:
    a = ll[-1] + ll[-2]
    ll.append(a)

ll.remove(ll[-1])   # remove the last number on the list since it it the first to be > 10000
print(ll)    
```

```python
# To make nice plots we import the library matplotlib.pyplot and call it plt
# This is just a simple plot, we will learn later how to make this look better
```

```python
import matplotlib.pyplot as plt
plt.plot(ll, color = 'red');
```

```python
# Take the sequence ll of Fib numbers, and make the sequence of the ratios of successive terms
# And then plot that sequence
```

```python
ratio = [ll[i+1] / ll[i] for i in range(len(ll)-1)]
import matplotlib.pyplot as plt
plt.plot(ratio, color = 'green', linestyle = '--');
```

# If, elif, else

```python
# Simple example
# (x % 2 == 0) is a Boolean expression wich returns True if x is even, and False if x is odd
l1 = [i for i in range(10)]
for x in l1:
    if x % 2 == 0:
        print(str(x) + ' is an even number')
    else:
        print(str(x) + ' is an odd number')
```

```python
# Simple example
# (x % 2 == 0) is a Boolean expression wich returns True if x is even, and False if x is odd

l1 = [i for i in range(10)]
for x in l1:
    if x % 2 == 0:
        print(str(x) + ' is an even number')
    else:
        print(str(x) + ' is an odd number')
```

```python
# Make this a little more complicated

l1 = [i for i in range(10)]
for x in l1:
    if x % 3 == 0:
        print(str(x) + ' is divisble by 3')
    elif x % 3 == 1:
        print(str(x) + ' is 1 plus a multiple of 3')
    else:
        print(str(x) + ' is 2 plus a multiple of 3')
```

## Exercise 2
Generate the first 100 terms for the sequence $x_n = Ax_n(1-x_n)$ for $A=2$, then $A=3$, $A=3.1$ etc... starting at $x_0=2$, and plot each sequence to observe that its "behavior" depends greatly on the choice of $A$.
Look here: https://en.wikipedia.org/wiki/Logistic_map

```python
l = [2, 3, 3.1]
for i in range(100):
    a = l[-1]*(1-l[-1])
    l.append(a)

print(l)
```

```python
import matplotlib.pyplot as plt 

l = [2, 3, 3.1]
for i in range(100):
    a = l[-1]*(1-l[-1])
    l.append(a)
plt.plot(l)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate the sequence for a given A
def generate_sequence(A, x0, n_terms):
    sequence = [x0]
    for _ in range(n_terms - 1):
        x_next = A * sequence[-1] * (1 - sequence[-1])
        sequence.append(x_next)
    return sequence

# Values of A to consider
A_values = [2, 3, 3.1]  # Add more values if needed

# Number of terms in the sequence
n_terms = 100

# Initial value of x0
x0 = 0.2

# Generate and plot the sequences for each value of A
for A in A_values:
    sequence = generate_sequence(A, x0, n_terms)
    plt.plot(sequence, label=f"A={A}")

# Plot settings
plt.xlabel('n (Term Index)')
plt.ylabel('x_n (Sequence Value)')
plt.title('Sequence Behavior for Different Values of A')
plt.legend()
plt.grid(True)
plt.show()
```
